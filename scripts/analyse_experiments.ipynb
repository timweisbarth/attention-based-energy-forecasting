```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
benchmark_table = pd.read_csv("./../results/benchmark_table_exp0.csv", header=[0, 1], index_col=[0, 1])

benchmark_table
```

```python
bm_load = benchmark_table.xs('load', level=0, axis=0)
bm_multi = benchmark_table.xs('multi', level=0, axis=0)

bm_load_mse = bm_load.xs('MSE', level=1, axis=1)
bm_load_mae = bm_load.xs('MAE', level=1, axis=1)

bm_multi_mae = bm_multi.xs('MAE', level=1, axis=1)
bm_multi_mse = bm_multi.xs('MSE', level=1, axis=1)

bm_load_mae

```

```python
fig, ax = plt.subplots(4, 1, figsize=(10, 10))

#for i in range(2):
#    ax[i].bar(bm_load, kind='bar', ax=ax[0], title='MSE')
bm_load_mse.plot(ax=ax[0], title='load: MSE')
bm_load_mae.plot(ax=ax[1], title='load: MAE')
bm_multi_mse.plot(ax=ax[2], title='multi: MSE')
bm_multi_mae.plot(ax=ax[3], title='multi: MAE')

```
