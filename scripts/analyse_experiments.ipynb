```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
benchmark_table = pd.read_csv("./../results/benchmark_table_exp0.csv", header=[0, 1], index_col=[0, 1])

benchmark_table
```

```python
bm_load = benchmark_table.xs('load', level=0, axis=0)
bm_load_mse = bm_load.xs('MSE', level=1, axis=1)
bm_load_mae = bm_load.xs('MAE', level=1, axis=1)
bm_load_mae

```

```python
fig, ax = plt.subplots(2, 1, figsize=(10, 5))

bm_load_mse.plot(kind='bar', ax=ax[0], title='MSE')
bm_load_mae.plot(kind='bar', ax=ax[0], title='MAE')
```
